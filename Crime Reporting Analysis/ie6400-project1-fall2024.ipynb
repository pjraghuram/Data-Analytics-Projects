```python
# remove duplicate rows
df.drop_duplicates(inplace=True)
df.shape
```

```python
# find all the columns which have empty or nan values
nullCols=df.apply(lambda col: col.isnull().sum()>0)
emptyColumnNames,emptyColumns=[],[]
for index, value in nullCols.items():
    if value == True:
        emptyColumns.append((index,df[index].dtype))
        emptyColumnNames.append(index)
emptyColumnNames
```

```python
#descriptive statistics of empty columns
descriptive_statistics=df[emptyColumnNames].describe(include='all')
descriptive_statistics
```

```python
# seperate the numerical, other type columns

numericEmptyColumns=df[emptyColumnNames].select_dtypes(include=["int64","float64"]).columns
otherEmptyColumns=df[emptyColumnNames].select_dtypes(exclude="number").columns
numericEmptyColumns,otherEmptyColumns
```

```python
# replace the numerical columns nan values with mean
df[numericEmptyColumns]=df[numericEmptyColumns].fillna(df[numericEmptyColumns].mean())
# replace the other  columns nan values with mode
df[otherEmptyColumns]=df[otherEmptyColumns].fillna(descriptive_statistics[otherEmptyColumns].loc['top'])
```

# 10. Predicting Future Trends:
• Employ time series forecasting methods, such as ARIMA or Prophet, to predict 
future crime trends based on historical data. Consider incorporating relevant 
external factors into your models

```python
import pandas as pd
from prophet import Prophet
import matplotlib.pyplot as plt


dateWiseCrimes=df.groupby('DATE OCC').size().reset_index(name='crimerate')
dates_df=dateWiseCrimes[['DATE OCC','crimerate']]
# Rename columns for Prophet
dates_df.columns = ['ds', 'y']

split_date="2024-01-01"
train=dates_df.loc[dates_df.ds<split_date].copy()
test=dates_df.loc[dates_df.ds>=split_date].copy()

train
```

```python
# Initialize Prophet model
model = Prophet()

# Fit the model with the data
model.fit(dates_df)
```

```python
#test Data


# Make predictions
forecast = model.predict(test)

# Plot the forecast
model.plot(forecast)
plt.title('Crime Data for test Data ')
plt.xlabel('Date')
plt.ylabel('Number of Crimes')
plt.show()
```

```python

# Create a DataFrame with future dates (e.g., next 6 months)
future = model.make_future_dataframe(periods=6, freq='M')

# Make predictions
forecast = model.predict(future)

# Plot the forecast
model.plot(forecast)
plt.title('Crime Forecast')
plt.xlabel('Date')
plt.ylabel('Number of Crimes')
plt.show()

# Optional: Print forecasted data
print(forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail())
```
